```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data
Fruits = pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Fruits.csv")
Fruits.head()
```

```python
Fruits.info()
```

```python
Fruits.describe()
```

```python
Fruits.columns
```

```python
Fruits.drop(('Fruit Name'),axis=1)
```

```python
y = Fruits['Fruit Category']
X = Fruits[['Fruit Weight','Fruit Width','Fruit Length','Fruit Colour Score']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
